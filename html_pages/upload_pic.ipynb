# Upload pic

1. define file
2. add timestamp
3. upload to place


```python
### imports
import os
import datetime
import requests
```

```python
### settings

file_name="test_image.png"
destination_url = 'https://cernbox.cern.ch/s/WlFpUqiA71ChjNs'
```

```python
### upload file

files = {'file': open(file_name, 'rb')}  # Specify the file you want to upload
response = requests.post(destination_url, files=files)

print(response.text)
```
